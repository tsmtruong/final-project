```python
## Import Dependencies and Data
import pandas as pd
import numpy as np
pd.set_option('mode.chained_assignment', None)

incomedf = pd.read_csv("Resources/RAW/San_Diego_Income_Data.csv", header=1)
rentdf = pd.read_csv("Resources/RAW/San_Diego_Financial_Characteristics.csv", header=1)
```

```python
incomedf
```

```python
## Drop any NaN and Margin of Error Columns
incomedf.dropna(axis=1, how='any', inplace=True)
incomedf.drop(list(incomedf.filter(regex = "Margin\sof\sError.")), axis= 1, inplace= True)
incomedf.drop(['Geography'], axis=1, inplace= True)
incomedf
```

```python
rentdf
```

```python
## Drop any NaN and Margin of Error Columns
rentdf.dropna(axis=1, how='any', inplace=True)
rentdf.drop(list(rentdf.filter(regex = "Margin\sof\sError.")), axis= 1, inplace= True)
rentdf.drop(['Geography'], axis=1, inplace= True)
rentdf
```

```python
sandiegohousingdf = incomedf.merge(rentdf, how='left', on='Geographic Area Name')
sandiegohousingdf
```

```python
#Make some Adjustments to the column names
sandiegohousingdf.columns = sandiegohousingdf.columns.str.replace('!!', '_')
sandiegohousingdf.columns = sandiegohousingdf.columns.str.replace('Estimate_', '')
sandiegohousingdf["Geographic Area Name"] = sandiegohousingdf["Geographic Area Name"].map(lambda x: x.lstrip('ZCTA5 '))

# Drop Columns with no data or little data
sandiegohousingdf.replace(to_replace='-', value=np.nan, inplace=True)
sandiegohousingdf.replace(to_replace='(X)', value=np.nan, inplace=True)
sandiegohousingdf = sandiegohousingdf.loc[sandiegohousingdf['Households_Total'] > 800]
sandiegohousingdf.dropna(axis=1, how='all', inplace=True)
sandiegohousingdf.drop(list(sandiegohousingdf.filter(regex = "Married-couple families_Mean income.")), axis= 1, inplace= True)

# Drop columns that are not needed
sandiegohousingdf.drop(list(sandiegohousingdf.filter(regex = "Percent.")), axis= 1, inplace= True)
sandiegohousingdf.drop(list(sandiegohousingdf.filter(regex = "Renter.")), axis= 1, inplace= True)
sandiegohousingdf.drop(list(sandiegohousingdf.filter(regex = "Owner.")), axis= 1, inplace= True)
sandiegohousingdf.drop(list(sandiegohousingdf.filter(regex = "Occupied.")), axis= 1, inplace= True)
sandiegohousingdf.drop(108, axis=0, inplace= True)

# Convert datatypes
for column in sandiegohousingdf:
    sandiegohousingdf[column] = sandiegohousingdf[column].astype('float')
sandiegohousingdf['Geographic Area Name'] = sandiegohousingdf['Geographic Area Name'].astype('int64')
sandiegohousingdf.rename(columns={'Geographic Area Name':'Zip Code'}, inplace=True)
sandiegohousingdf.set_index("Zip Code", inplace=True)


sandiegohousingdf
```

```python
householddf = sandiegohousingdf.filter(regex="Households_Total.").multiply(sandiegohousingdf["Households_Total"],axis=0)/100
houserounddf = householddf.round(decimals=0)
familydf = sandiegohousingdf.filter(regex="Families_Total.").multiply(sandiegohousingdf["Families_Total"],axis=0)/100
familyrounddf = familydf.round(decimals=0)
marrieddf = sandiegohousingdf.filter(regex="Married-couple families_Total.").multiply(sandiegohousingdf["Married-couple families_Total"],axis=0)/100
marriedrounddf = marrieddf.round(decimals=0)
nonfamilydf = sandiegohousingdf.filter(regex="Nonfamily households_Total.").multiply(sandiegohousingdf["Nonfamily households_Total"],axis=0)/100
nonfamilyrounddf = nonfamilydf.round(decimals=0)
```

```python
mergedf = houserounddf.join(familyrounddf, on='Zip Code', how='left').join(marriedrounddf, on='Zip Code', how='left').join(nonfamilyrounddf, on='Zip Code', how='left')
sandiegohousingdf.update(mergedf)
sandiegohousingdf
```

```python
from pathlib import Path  
filepath = Path('Resources/Clean/San_Diego_Housing_Data.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
sandiegohousingdf.to_csv(filepath)
```
