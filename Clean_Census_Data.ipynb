```python
## Import Dependencies and Data
import pandas as pd

incomedf = pd.read_csv("Resources/RAW/San_Diego_Income_Data.csv", header=1)
rentdf = pd.read_csv("Resources/RAW/San_Diego_Financial_Characteristics.csv", header=1)
```

```python
incomedf
```

```python
## Drop any NaN and Margin of Error Columns
incomedf.dropna(axis=1, how='any', inplace=True)
incomedf.drop(list(incomedf.filter(regex = "Margin\sof\sError.")), axis= 1, inplace= True)
incomedf.drop(['Geography'], axis=1, inplace= True)
incomedf
```

```python
rentdf
```

```python
## Drop any NaN and Margin of Error Columns
rentdf.dropna(axis=1, how='any', inplace=True)
rentdf.drop(list(rentdf.filter(regex = "Margin\sof\sError.")), axis= 1, inplace= True)
rentdf.drop(['Geography'], axis=1, inplace= True)
rentdf
```

```python
sandiegohousingdf = incomedf.merge(rentdf, how='left', on='Geographic Area Name')
sandiegohousingdf
```

```python
#Make some Adjustments to the column names

sandiegohousingdf.columns = sandiegohousingdf.columns.str.replace('!!', '_')
sandiegohousingdf.columns = sandiegohousingdf.columns.str.replace('Estimate_', '')

sandiegohousingdf.head(20)
```

```python
from pathlib import Path  
filepath = Path('Resources/Clean/San_Diego_Housing_Data.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
sandiegohousingdf.to_csv(filepath)
```
