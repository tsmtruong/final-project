```python
# importing dependencies
import pandas as pd
import glob
import os
```

```python
# setting path
files = os.path.join("Resources/RAW/Schools/greatschools*.csv")
```

```python
#list of merged files returned
files = glob.glob(files)
```

```python
# joining files 
school_df = pd.concat(map(pd.read_csv, files), ignore_index=True)

school_df
```

```python
# Separate Zpcodes into a new column and rename student count column 
school_df['zipcode'] = school_df['address'].str[-5:]
school_df['student_count'] = school_df['tablescraper-selected-row 3']

# Drop columns with irrelevant data
school_df.drop(['rating-circle-small','tablescraper-selected-row 2','tablescraper-selected-row 3','tablescraper-selected-row 4','tablescraper-selected-row href','tablescraper-selected-row 6','tablescraper-selected-row href 2','name href','tablescraper-selected-row','address'],axis=1,inplace=True)
school_df = school_df.set_index("zipcode")
school_df.dropna(axis=0,how='any',inplace=True)

# Convert datatypes
school_df['student_count'] = school_df['student_count'].astype('int64')


school_df
```

```python
# exporting file
from pathlib import Path

filepath = Path("Resources/Clean/schools.csv")
school_df.to_csv(filepath)
```
