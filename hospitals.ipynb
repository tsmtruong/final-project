```python
# import dependencies
import pandas as pd
import os
import glob
```

```python
# setting path
files = os.path.join("Resources/hospitals/ushospitalfinder*")

#list of merged files returned
files = glob.glob(files)

# joining files
hospital_df = pd.concat(map(pd.read_csv, files), ignore_index=True)

hospital_df.head(10)
```

```python
#print column names
print(hospital_df.keys())
```

```python
#dropping columns
clean_df = hospital_df.drop(['even','even 2'], axis=1, inplace=True)

hospital_df.head()
```

```python
# dropping duplicates
hosp_df = hospital_df.drop_duplicates(subset=['hospital_name'])

hosp_df.head()
```

```python
# extracting zipcodes

hosp_df['zipcode'] = hosp_df['address'].str[-10:]

hosp_df
```

```python
# take the last block of strings by splitting on a space and taking the last part
hosp_df['zipcodes'] = [x.split(' ')[-1] for x in hosp_df['address']]
# Check if any of them contains a -
hosp_df.loc[hosp_df['zipcodes'].str.contains('-')]
```

```python
hosp_df['first_zip'] = [x.split('-')[0] for x in hosp_df['zipcodes']]
```

```python
hosp_df.set_index('first_zip',inplace=True)
```

```python
# exporting file
from pathlib import Path

filepath = Path("Resources/Clean/hospitals.csv")
hosp_df.to_csv(filepath)
```
