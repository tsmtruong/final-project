```python
# importing dependencies
import pandas as pd
import numpy as np
```

```python
# read in csv file

housing_df = pd.read_csv("Resources/Clean/San_Diego_Housing_Data.csv")
housing_df
```

```python
#changing names

housing_df.columns = housing_df.columns.str.replace('!!', '_')

housing_df.head(20)
```

```python
flipped_df = housing_df.transpose()

flipped_df.head(30)
```

```python
#new header
new_header = flipped_df.iloc[1]
flipped_df = flipped_df[2:]
flipped_df.columns = new_header

flipped_df.head(30)
```

```python
# renaming columns

flipped_df.columns = flipped_df.columns.str.replace('ZCTA5', '')

flipped_df.head(30)
```

```python
# filtering the data

filtered_df = flipped_df[flipped_df['San Diego County, California'] != '(X)']

filtered_df.head(20)
```

```python
filtered_df = filtered_df.drop([' 92135', ' 92140', ' 92147', ' 92155'], axis=1)

filtered_df.head()
```
