```python
# import dependencies
import pandas as pd
```

```python
# Read in dataset
transit_data = pd.read_csv("Resources/Clean/San_Diego_Transit.csv", index_col=0)

# Convert dataset into DataFrame
transit_df = pd.DataFrame(transit_data)
transit_df
```

```python
# Rename the 'postcode' column to 'Zip Code'
transit_df = transit_df.rename(columns={'postcode' : 'Zip Code'})

# Reorder the columns
transit_df = transit_df[['Zip Code', 'stop_name', 'lat', 'lng', 'county']]

# Set the 'Zip Code' column as the index
transit_df = transit_df.set_index('Zip Code')
transit_df
```

```python
# exporting file
from pathlib import Path

filepath = Path("Resources/Clean/San_Diego_Transit.csv")
transit_df.to_csv(filepath)
```
