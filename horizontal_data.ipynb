```python
# import dependencies
import pandas as pd
import numpy as np

# read in csv
transposed_df = pd.read_csv('Resources/Clean/Transposed_Data.csv')
transposed_df.head()
```

```python
#transposing daataframe

horizontal_df = transposed_df.transpose()
horizontal_df.head()
```

```python
#new header
new_header = horizontal_df.iloc[0]
horizontal_df = horizontal_df[1:]
horizontal_df.columns = new_header

horizontal_df.head()
```

```python
from pathlib import Path
filepath = Path('Resources/Clean/Horizontal_Data.csv')
horizontal_df.to_csv(filepath)
```
