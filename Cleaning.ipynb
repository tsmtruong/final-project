```python
# Import dependencies
import pandas as pd
filepath = "Resources/Clean/San_Diego_Housing_Data.csv"

# Read in the Dataframe
df = pd.read_csv(filepath, index_col=0)
df.head()
```

```python
df.reset_index(inplace=True)
df = df.rename(columns = {'Geographic Area Name': 'Zipcode'})
df.head()
```

```python
# Drop 'San Diego CA, California' from zipcodes
df['Zipcode'].values


```

```python
# Remove SD from Zipcodes
new_df = df.drop(df.index[-1])
new_df.tail()
```

```python
# Remove Outliers
new_df = new_df.drop(new_df.index[97], axis=0)
new_df = new_df.drop(new_df.index[50], axis=0)
new_df = new_df.drop(new_df.index[18], axis=0)
```

```python
# Convert objects to floats (In hopes to transform testing data for machine learning model)
new_df['Estimate_Households_Median income (dollars)'] = new_df['Estimate_Households_Median income (dollars)'].astype('float')

new_df['Estimate_Families_Median income (dollars)'] = new_df['Estimate_Households_Median income (dollars)'].astype('float')

new_df['Estimate_Families_Mean income (dollars)'] = new_df['Estimate_Households_Median income (dollars)'].astype('float')
```
